### 네이버 쇼핑 사이트 크롤링하기

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://search.shopping.naver.com/best100v2/main.nhn')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('#popular_srch_lst > li > span.txt > a') 
for item in items:
    print(item.get_text())

# popular_srch_lst > li.on > span.txt > a      ->  리스트 하나만 나옴
# popular_srch_lst > li:nth-child(2) > span.txt > a
# 2순위를 카피 해보니 li. 다음이 다르구나 -> li만 남겨보자
```

```python
import requests
from bs4 import BeautifulSoup

site_list = ['https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000001','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000002','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000003']
for site in site_list:
    res = requests.get(site)
    soup = BeautifulSoup(res.content, 'html.parser')
    items = soup.select('#productListArea > ul > li > p > a') 
    print(site)
    for index, item in enumerate(items):
        print(str(index+1)+'.',item.get_text())
```

### Urillib 라이브러리 사용법
##### requests 라이브러리를 사용해서 크롤링을 진행하고, 문제가 있는 경우에만 urllib으로 실행해보면 좋음

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
res = urlopen('https://davelee-fun.github.io/')
soup = BeautifulSoup(res, 'html.parser')
items = soup.select('h4.card-text')
for item in items:
    print(item.get_text().strip())
```

#### http response code 크롤링할 페이지가 없는 경우 확인 방법
##### - requests 라이브러리의 경우, requests.get() 의 리턴변수에서 응답 코드 확인 가능
##### - http 규격에 따라 응답 코드가 200인 경우는 정상응답, 그렇지 않으면 문제가 있다는 뜻

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://davelee-fun.github.io/XXX')
if res.status_code != 200:
    print('페이지 없음')
else:
    soup = BeautifulSoup(res.content, 'html.parser')
    data = soup.select('h4.card-text')
    for item in data:
        print(item.get_text())
```

### 여러페이지 한번에 크롤링하는 기법

```python
import requests
from bs4 import BeautifulSoup

for page_num in range(10):
    if page_num ==0:
        res = requests.get('https://davelee-fun.github.io/')
    else:
        res = requests.get('https://davelee-fun.github.io/page'+str(page_num+1))
    soup = BeautifulSoup(res.content, 'html.parser')
    data=soup.select('h4.card-text')
    for item in data:
        print(item.get_text().strip())
```

```python
import requests
from bs4 import BeautifulSoup

print("상영작 예매순위")
for index, page_num in enumerate([107992,203488,191914,192620,187310,203097,196051,35546,203468,196215]):
    res = requests.get('https://movie.naver.com/movie/bi/mi/basic.nhn?code='+str(page_num))
    soup = BeautifulSoup(res.content, 'html.parser')
    data=soup.select('#content > div.article > div.mv_info_area > div.mv_info > h3 > a:nth-child(1)')
    index += 1
    for item in data:
        print(str(index)+'.', item.get_text().strip())
```

#### 크롤링해서 엑셀파일로 데이터 저장

```python
import openpyxl
def write_excel_template(filename, sheetname, listdata):
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    excel_sheet.column_dimensions['A'].width=100
    excel_sheet.column_dimensions['B'].width=20
    
    if sheetname != '':
        excel_sheet.title = sheetname
    for item in listdata:
        excel_sheet.append(item)
    excel_file.save(filename)
    excel_file.close()
```

```python
import requests
from bs4 import BeautifulSoup

product_lists=list()

for page_num in range(10):
    if page_num ==0:
        res = requests.get('https://davelee-fun.github.io/')
    else:
        res = requests.get('https://davelee-fun.github.io/page'+str(page_num+1))
    soup = BeautifulSoup(res.content, 'html.parser')
    data=soup.select('div.card')
    for item in data:
        product_name = item.select_one('div.card-body h4.card-text')
        product_date = item.select_one('div.wrapfooter span.post-date')
        product_info = [product_name.get_text().strip(), product_date.get_text()]
        product_lists.append(product_info)
```

```python
write_excel_template('tmp.xlsx','상품정보', product_lists)
```
